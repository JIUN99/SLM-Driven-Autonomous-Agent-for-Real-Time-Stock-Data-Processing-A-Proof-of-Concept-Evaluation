```python
import sqlite3

conn = sqlite3.connect('audit_log.db')   # open a connection to the file
cur = conn.cursor()                       # create a cursor to run queries

# Check the file isn't corrupted
cur.execute("PRAGMA integrity_check;")
print(cur.fetchall())

# List what tables exist inside
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cur.fetchall())

# Count rows
cur.execute("SELECT COUNT(*) FROM inference_audit_log;")
print(cur.fetchone())

# Look at the actual data
cur.execute("SELECT * FROM inference_audit_log LIMIT 5;")
for row in cur.fetchall():
    print(row)

conn.close()
```
